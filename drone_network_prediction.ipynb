### Data

#### supplied data

```python
# supplied data

df_stations = pd.read_csv('data/stations.csv')
df_services = pd.read_csv('data/services.csv')
df_drones = pd.read_csv('data/drones.csv')
df_schedules = pd.read_csv('data/schedules.csv')
df_actuals = pd.read_csv('data/actuals.csv')
```

```python
# datetime

df_schedules.scheduled_departure_datetime = df_schedules.scheduled_departure_datetime.astype('datetime64[ns]')
df_schedules.scheduled_arrival_datetime = df_schedules.scheduled_arrival_datetime.astype('datetime64[ns]')
df_actuals.actual_departure_datetime = df_actuals.actual_departure_datetime.astype('datetime64[ns]')
df_actuals.actual_arrival_datetime = df_actuals.actual_arrival_datetime.astype('datetime64[ns]')
```

```python
# Schedules

# waiting time at station (1.4)
df_schedules = df_schedules.assign(waiting = df_schedules.scheduled_departure_datetime
                                   - df_schedules.scheduled_arrival_datetime)
# different departures (1.4)
df_schedules = df_schedules.assign(departure_time = df_schedules.scheduled_departure_datetime.dt.time)

# different departures (1.4)
df_schedules = df_schedules.assign(departure_date = df_schedules.scheduled_departure_datetime.dt.date)
```

#### connections

```python
# connections
    
    # from station (dep) --> to station (arr)
    # includes different services, but not the schedule
    # alternative construction: duplicate columns and groupby + shift

df_departures = pd.merge(df_services, df_stations, on='station', sort=True)
df_arrivals = df_departures.copy()
df_departures.station_order = df_departures.station_order +1

df_connections = pd.merge(df_departures, df_arrivals, on=['service_id', 'station_order'],
                          sort=True, suffixes=('_dep', '_arr'), how='outer', indicator=True)
connections1 = df_connections[df_connections._merge == 'right_only'].dropna(axis=1)
connections2 = df_connections[df_connections._merge == 'left_only'].dropna(axis=1).drop(['station_order', '_merge'], axis=1)
connections3 = pd.merge(connections1, connections2, on=['service_id'], sort=True)
df_connections = df_connections[df_connections._merge == 'both'].append(connections3).sort_values(['service_id', 'station_order']).reset_index()
df_connections.rename(columns={'station_order': 'connection_order', 'type_dep': 'type'}, inplace=True)
df_connections.connection_order = df_connections.connection_order -1
df_connections.drop(['type_arr'], axis=1, inplace=True)
df_connections = df_connections.reindex(columns = ['service_id', 'connection_order', 'type', 'station_dep', 'continent_dep',
                                                   'country_code_dep', 'lat_lng_dep', 'major_city_dep', 'station_arr',
                                                   'continent_arr', 'country_code_arr', 'lat_lng_arr', 'major_city_arr'])
```

```python
# add variables

    # travel distance
    
df_connections = df_connections.assign(travel_distance = 0.0)
for index, row in df_connections.iterrows():
    df_connections.loc[index, 'travel_distance'] = geopy.distance.distance(row['lat_lng_dep'][1:-1],
                                                                           row['lat_lng_arr'][1:-1]).km
    
```

#### flights

```python
# flights
    
    # same as connections, but includes schedule
    # inlcudes schedules and services, but no detailed information about the station

    
df_flights = pd.merge(df_schedules, df_services, on=['service_id', 'station'])
df_flights = df_flights.sort_values(['service_id', 'scheduled_arrival_datetime']).reset_index(drop=True)

df_flights.scheduled_departure_datetime = df_flights.groupby('service_id').scheduled_departure_datetime.shift(1)
for i in ['station', 'forecast_windspeed_kts', 'station_order', 'station_call_id', 'waiting']:
    col = i + '_dep'
    df_flights = df_flights.assign(cols=df_flights.groupby('service_id')[i].shift(1))
    df_flights.rename(columns={'cols': col}, inplace=True)

df_flights.rename(columns={'station': 'station_arr', 'forecast_windspeed_kts': 'forecast_windspeed_kts_arr',
                           'station_order': 'station_order_arr', 'station_call_id': 'station_call_id_arr',
                           'waiting': 'waiting_arr'}, inplace=True)
df_flights = df_flights.reindex(columns = ['service_id', 'station_dep', 'station_order_dep', 'scheduled_departure_datetime',
                                           'station_arr', 'station_order_arr', 'scheduled_arrival_datetime', 
                                           'forecast_windspeed_kts_dep', 'waiting_dep', 'station_call_id_dep',
                                           'forecast_windspeed_kts_arr', 'waiting_arr', 'station_call_id_arr'])

df_flights.dropna(inplace=True)
    #df_flights = df_flights[df_flights.station_order_dep < df_flights.station_order_arr] in case services are not looped

df_flights.station_order_dep = df_flights.station_order_dep.astype('int')
```

```python
# add variables
    
    # flighttime (from departure to arrival station)
df_flights = df_flights.assign(flighttime = (df_flights.scheduled_arrival_datetime - df_flights.scheduled_departure_datetime).dt.seconds/60)
```

#### comprehensive

```python
# exchange variables

    # flighttime (flights --> connections)
df_connections = pd.merge(df_connections, df_flights[['service_id', 'station_arr', 'flighttime']].drop_duplicates(),
                      on=['service_id', 'station_arr'], how='left')

    # traveldistance (connections --> flights)
df_flights = pd.merge(df_flights, df_connections[['service_id', 'station_arr', 'travel_distance']],
                      on=['service_id', 'station_arr'])

    # average flight time for connection of two stations
avg_flighttime = pd.DataFrame(df_flights.groupby(['station_arr', 'station_dep']).mean().flighttime).reset_index()
df_connections = pd.merge(df_connections, avg_flighttime, on=['station_arr', 'station_dep'], suffixes=('' , '_avg'))
```

#### delays

```python
# add possible variables to actual departures and arrivals to predict delay factors

df_delays = pd.merge(df_actuals, df_stations[['station', 'lat_lng', 'country_code']], on='station')
df_delays = pd.merge(df_delays, df_services[['service_id', 'station', 'type']], on=['service_id', 'station'])
df_delays = pd.merge(df_delays, df_drones, on='service_id')
df_delays = pd.merge(df_delays, df_connections[['service_id', 'station_arr', 'flighttime', 'travel_distance']],
                left_on=['service_id', 'station'], right_on=['service_id', 'station_arr']).drop('station_arr', axis=1)
```

```python
# construct variables

    # date and time
df_delays = df_delays.assign(date = df_delays.actual_arrival_datetime.dt.date)
df_delays = df_delays.assign(time = df_delays.actual_arrival_datetime.dt.hour + df_delays.actual_arrival_datetime.dt.minute/60)

    # latitude
df_delays = df_delays.assign(lat = df_delays.lat_lng.apply(lambda x: float(x[1:].split(', ')[0])))
```

```python
# delay dummy

    # A flight is considered as delayed if the actual arrival is 15 minutes past the scheduled arrival
    # (according to FAA for airlines, I will use the same standard)

df_delays = df_delays.assign(delayed = df_delays.arrival_delay_seconds.apply(lambda x: 1 if (x/60>=15) else 0))
```

##### with dummies

```python
df_delays2 = pd.concat([df_delays, pd.get_dummies(df_delays.manufacturer)], axis=1)
df_delays2 = pd.concat([df_delays2, pd.get_dummies(df_delays2.type)], axis=1)
df_delays2 = pd.concat([df_delays2, pd.get_dummies(df_delays2.country_code)], axis=1)
df_delays2.columns = df_delays2.columns.str.replace(' ', '_').str.lower()
```

#### predicted delays

```python

df_delays_pred = pd.merge(df_delays[['station_call_id', 'arrival_delay_seconds', 'delayed']], df_schedules, on='station_call_id', how='right')
df_delays_pred = pd.merge(df_delays_pred, df_stations[['station', 'lat_lng', 'country_code']], on='station')
df_delays_pred = pd.merge(df_delays_pred, df_services[['service_id', 'station', 'type']], on=['service_id', 'station'])
df_delays_pred = pd.merge(df_delays_pred, df_drones, on='service_id')
df_delays_pred = pd.merge(df_delays_pred, df_connections[['service_id', 'station_arr', 'flighttime', 'travel_distance']],
                left_on=['service_id', 'station'], right_on=['service_id', 'station_arr']).drop('station_arr', axis=1)
df_delays_pred = df_delays_pred.assign(lat = df_delays_pred.lat_lng.apply(lambda x: float(x[1:].split(', ')[0])))

df_delays_pred = df_delays_pred.assign(service_num = df_delays_pred.service_id.astype('category').cat.codes)
df_delays_pred = df_delays_pred.assign(station_num = df_delays_pred.station.astype('category').cat.codes)
df_delays_pred = df_delays_pred.assign(arrival_time = df_delays_pred.scheduled_arrival_datetime.dt.hour + df_delays_pred.scheduled_arrival_datetime.dt.minute/60)
```
